```python
import pandas as pd 
df= pd.read_csv("E0.csv")
print(f"🚀 Dimensión del Dataset: {df.shape[0]} partidos y {df.shape[1]} variables distintas.")
columnas_clave = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'B365H', 'B365D', 'B365A']
df_filtrado = df[columnas_clave]
df_filtrado
```

```python
df.info()
```

```python
# hacemos el codigo legible 
traduccion = {'Date': 'Fecha',
    'HomeTeam': 'Equipo_Local',
    'AwayTeam': 'Equipo_Visitante',
    'FTHG': 'Goles_Local (FTHG)',
    'FTAG': 'Goles_Visitante (FTAG)',
    'FTR': 'Resultado_Final (FTR)',
    'B365H': 'Cuota_Local (B365H)',
    'B365D': 'Cuota_Empate (B365D)',
    'B365A': 'Cuota_Visitante (B365A)'
}
df_final = df_filtrado.rename(columns=traduccion)
print("columnas renombradas con exito")
df_final

 
```

```python
df_final ["Fecha"] = pd.to_datetime(df_final["Fecha"],format="%d/%m/%Y")
```

```python
import urllib
from sqlalchemy import create_engine

# 1. Credenciales y nombre de la tabla
server = 'DESKTOP-08BV2N8'
database = 'Betting_Project'
nombre_tabla_sql = 'Premier_Historico'

# 2. Creamos el Motor con tu Driver moderno (ODBC Driver 17)
parametros = urllib.parse.quote_plus(f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;")
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={parametros}")

# 3. La inyección masiva
print(f"⏳ Inyectando {df_final.shape[0]} partidos usando ODBC 17...")

df_final.to_sql(
    name=nombre_tabla_sql, 
    con=engine, 
    if_exists='append', 
    index=False         
)

print(f"✅ ¡Éxito total! Datos guardados en la tabla: {nombre_tabla_sql}")
```
